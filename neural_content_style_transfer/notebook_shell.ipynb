```python
from sklearn.model_selection import ParameterGrid
    
# zu betrachende Parameter definieren
param_grid = {
            'style_layers_and_style_weights': [{
            "conv1_1": 1.0,
            "conv2_1": 0.75,
            "conv3_1": 0.2,
            "conv4_1": 0.2,
            "conv5_1": 0.2,
        },
        {
            "conv1_1": 1.0,
            "conv2_1": 0.75,
            "conv4_1": 0.2,
        },
         {
            "conv1_1": 1.0,
            "conv1_2": 1.0,
            "conv2_1": 0.75,
            "conv2_2": 0.75,
            "conv3_1": 0.2,
            "conv4_1": 0.2,
            "conv5_1": 0.2,
        }],
            'content_layer': ["conv4_2","conv5_2","conv5_1"],
            'content_fashion_weight': [1],
            'content_art_weight': [10,100,1000],
            'style_art_weight': [1e6,1e5,1e7],
            'steps' : [300],
            'learning_rate' :[1000],
            'show_every': [100],
            'white_canvas': [True, False],
            'fashion_image_feature_is_white': [False, True]
    }

# Paramater Kombinationen durch ParamterGrid() kreieren, als liste gespeichert
param_grid = ParameterGrid(param_grid)


#rauschneiden von 100 Paramter Kombinationen großen Batches (eg. bei 216 Paramter Kombinationen 3 Batches a 100,100,16)
batches_param_comb =[]
batch_param_comb =[]
for ii in range(len(param_grid)):
  batch_param_comb.append(param_grid[ii])
  if (ii+1) % 100 == 0:
    batches_param_comb.append(batch_param_comb)
    batch_param_comb = []
  if ii == len(param_grid)-1:
    batches_param_comb.append(batch_param_comb)

# zum suchen später
#index =[]
#for i, param_combination in enumerate(param_grid):
 # if param_combination["content_layer"] == "conv4_2":
 #   index.append(i)

for batches in batches_param_comb:
  print(len(batches))
    

```

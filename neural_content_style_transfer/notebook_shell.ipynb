```python
#User definieren
User = "Daniel"
grenze_batch_einteilung = math.floor(len(batches_param_comb)/2)

if User == "Daniel":
  batches_param_comb = [batches for i, batches in enumerate(batches_param_comb) if i < grenze_batch_einteilung ]
  for i, batches in enumerate(tqdm(batches_param_comb)):
    for ii, param_combination in enumerate(batches):
      NCTS = NCTS_adapted_for_GS.NCTS(param_combination)
      images = NCTS.perform_ncts()
      for iii, im in enumerate(images):
        plt.imsave(f"grid_search_images/batch{i}_param_comb{ii}_step{iii}.jpeg",im)

if User == "Lukas":
  batches_param_comb = [batches for i, batches in enumerate(batches_param_comb) if i >= grenze_batch_einteilung ]
  len(batches_param_comb) 
  for i, batches in enumerate(batches_param_comb):
    for ii, param_combination in enumerate(batches):
      NCTS = NCTS_adapted_for_GS.NCTS(param_combination)
      images = NCTS.perform_ncts()
      for iii, im in enumerate(images):
        plt.imsave(f"grid_search_images/batch{grenze_batch_einteilung+i}_param_comb{ii}_step{iii}.jpeg",im)

```
